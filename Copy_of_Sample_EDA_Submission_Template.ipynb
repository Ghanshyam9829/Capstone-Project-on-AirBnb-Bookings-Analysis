## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#We can remove those columns which are not important 

dataset.drop(['name','host_id','host_name'],axis=1,inplace=True)


#hear am not droping the 'id' Because he id is having all 48895 unique values , So id will be a unique identity by that we can get any particuler row value 




````

````python
#Now we can create the a data set for which price is less than 100 for perticuler customer who can afford only 100 price 

dataset1 = dataset[dataset['price'] <= 100]

dataset1.head()

#now we get a dataset which price is less than 100 
````

````python
#By using the data set we can get the room_type which are available for minimum one night

dataset2 = dataset[dataset["minimum_nights"] == 1]

dataset2.head(100)

#now dataset2 will provide us the room_type where a customer can stay for one night 
````

````python
#Now we can find out the dataset3 for airbnb host which is available 365 Days 
dataset3 = dataset[dataset["availability_365"] == 365]

dataset3.head()
````

````python
#Lets combine dataset1 and dataset2 and see what concusion we get
dataset4 = dataset[(dataset['price'] <= 100) & (dataset["minimum_nights"] == 1)]

dataset4.head()

#Now we are geeting those listings on Airbnb which are available minimum one night with price less than 100
````
